```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import accuracy_score
import time


dfdata = pd.read_csv("primepatient_dataset.csv")
dfdata.head()
```

```python
X = dfdata.drop('outcome', axis=1)
y = dfdata['outcome']

# Function to evaluate model with different dataset sizes
def evaluate_model(sizes=[100, 1000, 10000, 100000, 1000000]):
    results = []

    for size in sizes:
        if size > len(dfdata):
            print(f"Skipping size {size} as it exceeds available data")
            continue

        # Sample the data
        sample_indices = np.random.choice(len(dfdata), min(size, len(dfdata)), replace=False)
        X_sample = X.iloc[sample_indices]
        y_sample = y.iloc[sample_indices]

        # Split into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(X_sample, y_sample, test_size=0.2, random_state=42)

        # Create and train the model
        model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')

        # Measure training time
        start_time = time.time()

        # Perform 5-fold cross-validation
        cv_scores = cross_val_score(model, X_train, y_train, cv=5)

        # Fit the model on the entire training set
        model.fit(X_train, y_train)

        # Measure elapsed time
        elapsed_time = time.time() - start_time

        # Test on the test set
        y_pred = model.predict(X_test)
        test_accuracy = accuracy_score(y_test, y_pred)

        # Collect results
        results.append({
            'Dataset_Size': size,
            'CV_Accuracy': cv_scores.mean(),
            'Test_Accuracy': test_accuracy,
            'Training_Time_Seconds': elapsed_time
        })

        print(f"Completed evaluation for size {size}")

    # Create a results dataframe
    results_df = pd.DataFrame(results)
    return results_df

# Run the evaluation
results = evaluate_model()

# Display the results
print("\nResults:")
print(results)
```
